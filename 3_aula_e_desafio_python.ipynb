```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
```

```python
df_principal = pd.read_excel("acoes_pura.xlsx", sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("acoes_pura.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("acoes_pura.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("acoes_pura.xlsx", sheet_name="ChatGPT")
df_chatgpt
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Ativo':'ativo','Data':'data','Último (R$)':'valor_final','Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
# var_pct é a coluna a ser criada!
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal # Remover coluna 'Código'
```

```python
df_principal['Variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal # Criação de coluna 'Variacao_rs'
```

```python
pd.options.display.float_format = '{:.2f}'.format #transforma o número de notação científica para float
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal # Transformar visualização dos números para int
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal # Renomear coluna de 'Qtde. Teórica' para 'Qtd_teorica'
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal # utilização de If para criar a coluna Resultado, que determina se o valor Subiu, Desceu ou se manteve estável
```

```python
df_principal = df_principal.merge(df_ticker, left_on='ativo', right_on='Ticker', how='left')
df_principal # Vlookup para trazer a tabela 'Ticker'
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Empresa', how='left')
df_principal = df_principal.drop_duplicates(subset=['ativo'])
df_principal # Vlookup para trazer a tabela 'ChatGPT'
```

```python
df_principal = df_principal.rename(columns={'Tempo desde a criação (anos)':'Idade'}).copy()
df_principal # Renomear a coluna para 'Idade'
```

```python
df_principal['ano_de_fundacao'] = 2024 - df_principal['Idade'] # adicionar a coluna 'ano_de_fundacao'
df_principal
```

```python
df_principal = df_principal.drop(columns=['Nome']).copy() # Remover a coluna 'nome'
df_principal
```

```python
df_principal = df_principal.drop(columns=['Ticker']).copy() # Remover a coluna 'Ticker'
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal # Utilização de If para dizer qual categoria de idade se encaixa cada empresa
```

```python

# Calculando o máximo, mínimo e média geral
maior = df_principal['Variacao_rs'].max()
menor = df_principal['Variacao_rs'].min()
media_geral = df_principal['Variacao_rs'].mean()

# Calculando a média separadamente para os grupos 'Subiu' e 'Desceu'
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Exibindo os resultados e formatando como float
print(f"Maior\t{maior:,.2f}")
print(f"Menor\t{menor:,.2f}")
print(f"Média\t{media_geral:,.2f}")
print(f"Média de quem subiu\t{media_subiu:,.2f}")
print(f"Média de quem desceu\t{media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
# groupby somando 'Desceu', 'Subiu' e 'Estável'
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()

# Arredonda para duas casas decimais e adiciona separadores de milhares
df_analise_saldo['Variacao_rs_formatado'] = df_analise_saldo['Variacao_rs'].apply(lambda x: '{:,.2f}'.format(x))

# imprimir o data frame
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Variacao_rs', text='Variacao_rs_formatado', y='Resultado', title='Variação Reais por Resultado')
fig.show()
```

```python
# Criar o gráfico de torta (pie)
fig = px.pie(df_analise_segmento, values='Variacao_rs', names='Segmento', title='Distribuição do valor por Segmento')

# Personalizar a legenda dentro de cada fatia da pizza
fig.update_traces(textinfo='label+percent', hovertemplate='Segmento: %{label}<br>Valor: R$ %{value:,.2f}', textposition='inside')

# Imprimir o gráfico
fig.show()
```

```python
df_faixa_etaria = df_principal.groupby('Cat_idade')['Variacao_rs'].sum().reset_index()
df_faixa_etaria['Variacao_rs_formatado'] = df_faixa_etaria['Variacao_rs'].apply(lambda x: '{:,.2f}'.format(x))
df_faixa_etaria
```

```python
df_count_idade = df_principal['Cat_idade'].value_counts().reset_index()
df_count_idade = df_count_idade.rename(columns={'Cat_idade':'Qtd_de_empresas'}).copy()
df_count_idade = df_count_idade.rename(columns={'index':'Cat_idade'}).copy()
df_count_idade
```

```python
df_faixa_etaria['Cat_idade'] = df_faixa_etaria['Cat_idade'].astype(str)
df_count_idade['Cat_idade'] = df_count_idade['Cat_idade'].astype(str)

# vlookup para trazer o numero de quantidade de empresas para sua respectiva categoria de idade
df_faixa_etaria = df_faixa_etaria.merge(df_count_idade, left_on='Cat_idade', right_on='Cat_idade', how='left')

# formatação para melhor visualização
df_faixa_etaria['Variacao_rs_formatado'] = df_faixa_etaria['Variacao_rs'].apply(lambda x: '{:,.2f}'.format(x))

df_faixa_etaria
```

```python
fig = px.bar(df_faixa_etaria, x='Variacao_rs', text='Variacao_rs_formatado', y='Cat_idade', title='Variação (R$) de Empresas por Idade')
fig.show()
```

```python
  # Criar o gráfico de torta (pie)
fig = px.pie(df_count_idade, values='Qtd_de_empresas', names='Cat_idade', title='Distribuição de Empresas por faixa etária')

# Personalizar a legenda dentro de cada fatia da pizza
fig.update_traces(textinfo='label+percent', hovertemplate='Qtd_de_empresas: %{label}<br>%{value:} Unidades', textposition='auto')

# Imprimir o gráfico
fig.show()
```
